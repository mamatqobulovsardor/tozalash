```python
import pandas as pd
import numpy as np
```

```python
data = {
    "ID": [1, 2, 3, 4, 5],
    "Ism": ["Ali", "Zaynab", "Jamshid", "Nilufar", "Botir"],
    "Yoshi": [34, np.nan, 28, 45, 52],
    "Daromad ($)": [50000, 75000, 30000, np.nan, 90000],
    "Kredit Balansi ($)": [2000, np.nan, 500, 1000, 0],
    "Kredit Reytingi": [5, np.nan, 4, 3, 4],
    "Shahar": ["Tashkent", "Samarkand", "Namangan", "Tashkent", np.nan],
    "Mijoz Davomiyligi (Yillar)": [3, 5, 2, np.nan, 10]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#Yo'qolgan qiymatlarni aniqlash
df.isnull().sum()
```

```python
#Yoshi ustunini o'rtacha qiymat bilan tuldiramz
df['Yoshi'] = df['Yoshi'].fillna(df['Yoshi'].mean())
```

```python
#Daromadni o'rtacha qiymatini bilan tuldiramz
df['Daromad ($)'] = df['Daromad ($)'].fillna(df['Daromad ($)'].mean())
```

```python
#Shahar ustunini eng kup uchraydigan shahar nomi bilan tuldiramz
df['Shahar'] = df['Shahar'].fillna(df['Shahar'].mode()[0])
```

```python
df['Kredit Balansi ($)'] = df['Kredit Balansi ($)'].fillna(df['Kredit Balansi ($)'].mean())
```

```python
df['Kredit Reytingi'] = df['Kredit Reytingi'].fillna(df['Kredit Reytingi'].mean())
```

```python
#Mijoz Davomiyligi (Yillar) urtacha qiymati bilan to'lidirish
df['Mijoz Davomiyligi (Yillar)'] = df['Mijoz Davomiyligi (Yillar)'].fillna(df['Mijoz Davomiyligi (Yillar)'].mean())
```
